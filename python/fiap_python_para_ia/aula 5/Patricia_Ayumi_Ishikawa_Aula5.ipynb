```python
#Desafio 1
import numpy as np
from pandas import Series, DataFrame
import pandas as pd

data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat','dog', 'dog'],
       'age': [2.5, 3, 0.5, np.nan, 5,2, 4.5, np.nan, 7,3],
       'visits': [1,3,2,3,2,3,1,1,2,1],
       'priority': ['yes','yes','no','yes','no','no','no','yes','no','no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

#dataframe
df = DataFrame(data, index=labels)
print(df)

#preenchendo com a media
df = df.fillna(df.mean())
print(df)

#lista unica de animais
print(df.animal.unique())

#estat√≠stica
print(df.describe())
```

```python
#Desafio 2

def mult(df):
    df2 = df.copy()
    df2.visits = df.visits * 2
    return df2

print(df)

#Duplica coluna visit - Outro DataFrame
print(mult(df))

#Rank 
df['rank'] = df['visits'].rank(ascending=0)
print(df)

#New_animals
arrNewAnimals = ['cow', 'horse', 'shark']
df['new_animals'] = df.animal.isin(arrNewAnimals)
print(df)
```
