### 1) Leia o arquivo usando pandas, imprima as 10 primeiras linhas

```python
# resposta 1
import pandas as pd
import numpy as np
```

```python
poke = pd.read_csv('Pokemon.csv')
poke.head(n=10)
```

### 2) Altere o nome das colunas passando todas as letras para maiúscula. Substitua o "." por ""

```python
# resposta 2
poke.columns = poke.columns.str.upper().str.replace('.', '') #change into upper case
poke.head()
```

### 3) Imprima apenas os pokemons legendários

```python
# resposta 3

poke[poke['LEGENDARY'] == True]
```

### 4) Verifique os valores faltantes existentes no dataset

```python
# resposta 4

poke.isna().sum()
```

### 5) Preencha os valores faltantes da coluna TYPE 2 com os valores correspondentes da coluna TYPE 1

```python
# resposta 5
poke['TYPE 2'] = poke['TYPE 2'].fillna(poke['TYPE 1'])
poke.isna().sum()
```

### 6) Agrupe os pokemons por TYPE 1, ordene pela quantidade em ordem decrescente

```python
# resposta 6

poke.groupby(by = 'TYPE 1').size().sort_values(ascending =False)
```

### 7) Type 1 == Fire ou Type 2 == Dragon & Type 1 == Dragon ou Type 2 == Fire

```python
dataFilter = ((poke['TYPE 1'] == 'Fire') | (poke['TYPE 2'] == 'Dragon') & (poke['TYPE 1'] == 'Dragon') | (poke['TYPE 2'] == 'Fire'))
maximo = poke[dataFilter]['HP'].max()
pokemonsFiltrados = poke[dataFilter]
pokemonsFiltrados[poke['HP'] == maximo]
```
